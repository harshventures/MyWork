# Introduction

![](https://c.tenor.com/t4VS-Zxh8xgAAAAC/lilo-e-stitch-lilo-and-stitch.gif)

Picture Credit: https://c.tenor.com

**What is Churn rate?**
> Churn rate (sometimes called attrition rate), in its broadest sense, is a measure of the number of individuals or items moving out of a collective group over a specific period. It is one of two primary factors that determine the steady-state level of customers a business will support.
> 
> Derived from the butter churn, the term is used in many contexts but most widely applied in business with respect to a contractual customer base. Examples include a subscriber-based service model as used by mobile telephone networks and pay TV operators. The term is often synonymous with turnover, for example participant turnover in peer-to-peer networks. Churn rate is an input into customer lifetime value modeling, and can be part of a simulator used to measure return on marketing investment using marketing mix modeling.

We want to know why customers are leaving or why they keep serving. Knowing the reason will enable us to devise strategies to prevent customer churn in the future.

What should we do for this?
First, we create a good model to predict churn well. Then, it checks what features the model has determined to be important.
Using these features, we think it would be good to devise a strategy to prevent customer churn.

This notebook intends to proceed in the following order.
1. Perform anomaly detection to check the data determined to be outliers in the given dataset. If you are lucky, you may get some good insight during this process.
2. After designing and training the deep learning model, check what features are important from the perspective of deep learning.
3. Design a model using Classic ML and interpret the models from various perspectives. Through this, we check the important features from the classic ML perspective.

If we check the dataset from various viewpoints as above, we will be able to have some insight.

# Anomaly Dectection

![](http://3.bp.blogspot.com/-t0agOH8DMOE/USg6kdHohNI/AAAAAAAAACM/3XuR0x38vLA/s1600/gestalt-pattern-image[1].jpg)

Picture Credit: http://3.bp.blogspot.com

> A real-world dataset often contains anomalies or outlier data points. The cause of anomalies may be data corruption, experimental or human errors. The presence of anomalies may impact the performance of the model, hence to train a robust data science model, the dataset should be free from anomalies

Ref: https://towardsdatascience.com

```python
def plot_hist(df, feature):    
    fig = px.histogram(df, x=feature,
                       color="Churn", 
                       marginal="box",
                       barmode ="overlay",
                       histnorm ='density'
                      )  
    fig.update_layout(
        title_font_color="white",
        legend_title_font_color="yellow",
        title={
            'text': feature+" histogram",
            'x':0.5,
            'xanchor': 'center',
            'yanchor': 'top'},
        paper_bgcolor="black",
        plot_bgcolor='black',
        font_color="white"
    )
    fig.show()
```

```python
from pycaret.anomaly import *
```

```python
_ = pycaret.anomaly.setup(
    data=telco_df,
    silent=True)
```

```python
pca = pycaret.anomaly.create_model('pca')
```

```python
pca_df = pycaret.anomaly.assign_model(pca)
```

```python
abnormal_data = pca_df[pca_df.Anomaly == 1].sort_values(by='Anomaly_Score', ascending=False)
normal_data = pca_df[pca_df.Anomaly == 0].sort_values(by='Anomaly_Score', ascending=False)
print("the size of anomaly = ",len(abnormal_data))
abnormal_data.head().style.set_properties(**{'background-color': 'black',
                           'color': 'white',
                           'border-color': 'white'})
```

# Deep Learning using Fastai

![](https://www.nicepng.com/png/detail/136-1365352_feature-fast-and-easy-icon.png)

Picture Credit: https://www.nicepng.com/

```python
from fastai.tabular.all import *
from fastinference.tabular import *
sns.set(style="ticks", context="talk")
plt.style.use("dark_background")
```

-------------------------------------
## Making Pipeline and DataLoader

```python
categorical_feature = telco_df.dtypes==object
categorical_cols = telco_df.columns[categorical_feature].tolist()
categorical_cols.remove('Churn')

numeric_feature = telco_df.dtypes!=object
numerical_cols = telco_df.columns[numeric_feature].tolist()
```

```python
procs = [Categorify, FillMissing, Normalize]
splits = RandomSplitter(valid_pct=0.2)(range_of(telco_df))
```

```python
to = TabularPandas(telco_df,
                   procs, 
                   categorical_cols,
                   numerical_cols, 
                   y_names="Churn", 
                   splits=splits)
dls = to.dataloaders(bs=16)
```

```python
dls.show_batch()
```

------------------------------------------
## Modeling

```python
learn = tabular_learner(dls, 
                        layers=[100,200,100], 
                        metrics=accuracy,
                        cbs = [EarlyStoppingCallback(monitor='accuracy', min_delta=0.001, patience=2),ActivationStats(with_hist=True)]
                       )
learn.model
```

```python
learn.model.layers
```

## Embeddings for Categorical Variables

Embedding reduces memory usage and speeds up neural networks than one-hot encoding. In addition, it expresses the essential properties of categorical variables that are mapped close to each other between similar values of categorical variables in the embedding space. Embedding handles datasets of high-order cardinality features particularly well compared to other approaches that tend to overfit. Because embedding defines the distance measurement between categorical variables, it is also useful for visualization of categorical data and data clustering.

Also, embedding finds an appropriate value through learning.

```python
learn.model.embeds
```

--------------------------------------
## Training

![](https://jehyunlee.github.io/2020/03/28/Python-DL-3-1cycle-learning-rate-policy/lr_low-81d04746df552258b87b24e2e3da17a5.gif)

Among DL hyperparameters, learning rate is an important hyperparameter. We will find an appropriate learning rate using the API provided by fastai, and train using it.

```python
plt.rcParams["figure.figsize"] = (8,6)
sr = learn.lr_find()
sr.valley
```

```python
learn.fine_tune(1, sr.valley)
```

```python
learn.fit_one_cycle(100, sr.valley)
```

-------------------------------------------
## Checking Overfitting and Underfitting

```python
plt.rcParams["figure.figsize"] = (8,6)
learn.recorder.plot_loss()
```

----------------------------------------
## Checking Activation

We can check whether the learning is successful by observing the activation values of each layer.

```python
def plot_layer_stats(self, idx):
    plt,axs = subplots(1, 3, figsize=(15,3))
    plt.subplots_adjust(wspace=0.5)
    for o,ax,title in zip(self.layer_stats(idx),axs,('mean','std','% near zero')):
        ax.plot(o)
        ax.set_title(f"{-1*layer}th layer {title}")
```

```python
for layer in range(1,4):
    plot_layer_stats(learn.activation_stats,-1*layer)
```

```python
def color_dim(self, idx):
    with plt.rc_context({"figure.figsize": (10,40), "figure.dpi": (600)}):
        res = self.hist(idx)
        fig = plt.figure()
        ax = fig.add_subplot(111)
        ax.imshow(res, origin='lower')
        ax.set_title(f"{idx}th activation histogram")
        ax.axis('off')
```

```python
matplotlib.rcParams['image.cmap'] = 'rainbow_r'
for layer in range(1,4):
    color_dim(learn.activation_stats,-1*layer)
```

<span style="color:Blue"> Observation:
* As a result of observing the activation value, there does not seem to be a situation in which the distribution is concentrated to a specific value.

Picture Credit: https://t3.ftcdn.net

```python
def display_stat(feature):
    mean = telco_df[feature].mean()
    std = telco_df[feature].std()
    skew = telco_df[feature].skew()
    kurtosis = telco_df[feature].kurtosis()
    print('mean: {0:.4f}, std: {1:.4f}, skew: {2:.4f}, kurtosis: {3:.4f} '.format(mean, std, skew, kurtosis))
```

# Classic Machine Learning

```python
from pycaret.classification import *
sns.set(style="ticks", context="talk",font_scale = 1)
plt.style.use("dark_background")
```

--------------------------------------
## Making Pipeline

```python
_ = setup(data = telco_df, 
      target = 'Churn',
      numeric_features = all_cols,
      silent = True)
```

-------------------------------
## Comparing Models

```python
top4 = compare_models(sort='Accuracy',n_select = 4,exclude=['xgboost','dummy','dt','svm','qda','knn','nb','ridge'])
```

-------------------------------------------
## Creating Models

```python
gbc = create_model('gbc',verbose = False)
lr = create_model('lr',verbose = False)
ada = create_model('ada',verbose = False)
catboost = create_model('catboost',verbose = False)
dt = create_model('dt',verbose = False)
```

-----------------------------
## Tuning Hyperparameters

```python
tuned_gbc = tune_model(gbc, optimize = 'Accuracy',verbose = False,n_iter = 20)
tuned_lr = tune_model(lr, optimize = 'Accuracy',verbose = False)
tuned_ada = tune_model(ada, optimize = 'Accuracy',verbose = False)
tuned_catboost = tune_model(catboost, optimize = 'Accuracy',verbose = False,n_iter = 20)
```

```python
params = { "min_samples_leaf": [100],"max_depth":[3]}
tuned_dt = tune_model(dt,early_stopping=True,optimize='Recall',custom_grid = params)
```

---------------------------------------------
## Interpreting Models

```python
with plt.rc_context({'figure.facecolor':'lightgrey'}):
    interpret_model(tuned_catboost)
```

## Ensemble

![](https://pluralsight2.imgix.net/guides/27d18692-27b8-42aa-8aa5-9668e625c41c_1.jpg)

Picture Credit: https://pluralsight2.imgix.net

Whether human or machine, there can be problems with cognitive biases. If we interpret the same dataset from various viewpoints and combine the prediction results of each model to reach a final conclusion, not only the cognitive bias problem can be solved, but also the performance improvement and overfitting problems can be solved.
Therefore, in classic ML, ensemble is a very important modeling method.

```python
blend_soft = blend_models(estimator_list = [gbc,ada,lr,catboost], optimize = 'Accuracy',method = 'soft')
```

```python
final_model = finalize_model(blend_soft)
```
